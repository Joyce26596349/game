# 猜數字遊戲

```python
#include <iostream>
using namespace std;
int main()
{
    while(true){
        srand(time(NULL));
        int num = rand() % 100 + 1;
        int low = 1, high = 100;
        cout << "這是一個猜數字遊戲，跟據系統提示猜出正確數字吧" << endl;
        cout << "輸入0000可結束遊戲 輸入1111可結束此輪" << endl;
        int input;
        int count;
        count = 0;
        while (true){
            cout << "請輸入" << low << "~" << high << "之間的數" << endl;
            cin >> input;
            if (input == 0000){
                return 0;
            }
            if (input == 1111){
                break;
            }
            while (low > input || high < input){
                cout << "輸入錯誤" << endl;
                cout << "請輸入" << low << "~" << high << "之間的數" << endl;
                cin >> input;
            }
            count += 1;
            if (num > input){
                cout << "再大" << endl;
                low = input;
            }
            else if (num < input){
                cout << "再小" << endl;
                high = input;
            }
            else{
                cout << "恭喜猜中" << endl;
                cout << "你猜了" << count << "次" << endl;
                break;
            }
        }
        cout << "數字為:" << num << endl;
    }
}
```
